```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
import urllib.request, json
import requests
from requests import Request, Session
from requests.auth import HTTPBasicAuth
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
```

```python
import time
```

```python
tickerlist = pd.read_csv('mongoTickerPriceFetch.csv') 
```

```python
coinids = tickerlist['coinGeckoID'].tolist()
```

```python
result = list()
def getcoinfo(id):
    rr = requests.get('https://api.coingecko.com/api/v3/coins/'+id+'?tickers=true')
    data=json.loads(rr.text)
    time.sleep(1.5)
    return result.append(data)
```

```python
for coinid in coinids:
    getcoinfo(coinid)
```

```python
df = pd.DataFrame(pd.json_normalize(result))
```

```python
df_result = pd.concat([tickerlist, df], axis=1)
```
