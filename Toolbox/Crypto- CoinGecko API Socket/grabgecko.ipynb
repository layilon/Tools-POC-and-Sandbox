Define function to grab crypto market cap info

```python
def getmarkinfo(page):
    rr = requests.get('https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=250&page='+str(page)+'&sparkline=false')
    data=json.loads(rr.text)
    df= pd.DataFrame(pd.json_normalize(data))
    return df
```
