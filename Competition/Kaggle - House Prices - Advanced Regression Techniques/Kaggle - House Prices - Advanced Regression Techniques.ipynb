```python
import pandas as pd
```

```python
from pycaret.regression import *
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
submission = pd.read_csv('sample_submission.csv')
```

```python
test.info()
```

```python
s = setup(train, target = 'SalePrice', session_id = 123, feature_selection = True,feature_selection_threshold=0.06, remove_multicollinearity = True, multicollinearity_threshold = 0.6, imputation_type="iterative", categorical_iterative_imputer="lightgbm", numeric_iterative_imputer="lightgbm", iterative_imputation_iters = 5, ignore_low_variance=True)
```

```python
best = compare_models()
```

```python
catboost = create_model('catboost')
```

```python
tuned_catboost = tune_model(catboost, search_library = 'optuna', n_iter = 10)
```

```python
predictions = predict_model(tuned_catboost, data=test)
```

```python
predictions.head()
```

```python
predictions.to_csv('submission.csv', index = False)
```

```python
sub = predictions[['Id', 'Label']].rename(columns={"Label": "SalePrice"})
sub.to_csv('submission.csv', index = False)
```
