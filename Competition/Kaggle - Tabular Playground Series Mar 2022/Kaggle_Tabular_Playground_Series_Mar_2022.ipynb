```python
import pandas as pd
```

```python
import datetime as dt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from pycaret.regression import *
```

```python
train = pd.read_csv('train.csv',index_col='row_id')
test = pd.read_csv('test.csv',index_col='row_id')
```

```python
train['time'] = pd.to_datetime(train['time'])
train['hour'] = train['time'].apply(lambda x : x.hour)
train['weekday'] = train['time'].dt.weekday
test['time'] = pd.to_datetime(test['time'])
test['hour'] = test['time'].apply(lambda x : x.hour)
test['weekday'] = test['time'].dt.weekday
train['(x, y, direction)'] = '('+train['x'].astype(str) +',' +train['y'].astype(str) +','+ train['direction']+')'
test['(x, y, direction)'] = '('+test['x'].astype(str) +',' +test['y'].astype(str) +','+ test['direction']+')'
train.drop(['time', 'x', 'y', 'direction'], axis=1, inplace=True)
test.drop(['time', 'x', 'y', 'direction'], axis=1, inplace=True)
```

```python
le = LabelEncoder()
```

```python
train['(x, y, direction)'] = le.fit_transform(train['(x, y, direction)']) 
test['(x, y, direction)'] = le.transform(test['(x, y, direction)'])
```

```python
s = setup(train, target = 'congestion', session_id = 123)
```

```python
best = compare_models()
```

```python
dt = create_model('dt')
```

```python
predictions = predict_model(dt, data=test)
```

```python
sub = predictions['Label']
```
